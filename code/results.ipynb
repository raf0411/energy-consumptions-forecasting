```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Tn 	min temperature (°C)
# Tx 	max temperature (°C)
# Tavg 	avg temperature (°C)
# RH_avg 	avg humidity (%)
# RR 	rainfall (mm)
# ss 	duration of sunshine (hour)
# ff_x 	max wind speed (m/s)
# ddd_x 	wind direction at maximum speed (°)
# ff_avg 	avg wind speed (m/s)
# ddd_car 	most wind direction (°)
# station_id 	station id which record the data. Detail of the station can be found in station_detail.csv
```

```python
ann_rmse = 0.006550738442623335 # Replace with your actual RMSE value for ANN
cnn_rmse = 0.006492818401602504 # Replace with your actual RMSE value for CNN
rnn_rmse = 0.006902055914377307  # Replace with your actual RMSE value for RNN
lstm_rmse = 0.00675362571655609  # Replace with your actual RMSE value for LSTM

ann_r2 = 0.781958808794495 # Replace with your actual R2 score for ANN
cnn_r2 = 0.7857974977518037  # Replace with your actual R2 score for CNN
rnn_r2 = 0.7579444895612493  # Replace with your actual R2 score for RNN
lstm_r2 = 0.7682434568019331  # Replace with your actual R2 score for LSTM

# Create a DataFrame
results = {
    'Model': ['ANN', 'CNN', 'RNN', 'LSTM'],
    'RMSE': [ann_rmse, cnn_rmse, rnn_rmse, lstm_rmse],
    'R2 Score': [ann_r2, cnn_r2, rnn_r2, lstm_r2]
}
results_df = pd.DataFrame(results)

# Display the DataFrame
print(results_df)

# Bar chart for RMSE
plt.figure(figsize=(12, 6))
plt.bar(results_df['Model'], results_df['RMSE'], color='darkgreen')
plt.title('Comparison of RMSE for Different Models')
plt.xlabel('Model')
plt.ylabel('RMSE')
plt.show()

# Bar chart for R2 Score
plt.figure(figsize=(12, 6))
plt.bar(results_df['Model'], results_df['R2 Score'], color='darkblue')
plt.title('Comparison of R2 Score for Different Models')
plt.xlabel('Model')
plt.ylabel('R2 Score')
plt.show()
```
